```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
traffic = pd.read_csv('tomtom_traffic.csv')
eco = pd.read_csv('oecd_city_economy.csv')
```

```python
# mostrar las primeras 5 filas de traffic
traffic.head()
```

```python
# mostrar las primeras 5 filas de eco
eco.head()
```

```python
# Examinar la estructura de traffic
traffic.info()
```

```python
# Examinar la estructura de eco
eco.info()
```

```python
# Estandarizar los nombres de las columnas de traffic
traffic.columns = traffic.columns.str.lower().str.replace(' ', '_')

# Verificar que los cambios se hayan aplicado
traffic.info()
```

```python
# estandarizar los nombres de las columnas de eco
eco.columns = eco.columns.str.lower().str.replace(' ', '_')

# Verificar los cambios
eco.info()
```

```python
# 1. Limpiar y convertir columnas en un solo paso por columna
eco['city_gdp_per_capita'] = pd.to_numeric(eco['city_gdp_per_capita'].astype(str).str.replace('.', '').str.replace(',', '.'), errors='coerce')
eco['unemployment_pct'] = pd.to_numeric(eco['unemployment_pct'].astype(str).str.replace('%', '').str.replace(',', '.'), errors='coerce')
eco['population_m'] = pd.to_numeric(eco['population_m'].astype(str).str.replace(',', '.'), errors='coerce')

# 2. Crear la población total
eco['population'] = eco['population_m'] * 1000000

# 3. Mostrar resultado
eco.info()
eco.head(3)
```

```python
traffic['updatetimeutc'] = pd.to_datetime(traffic['updatetimeutc'], errors='coerce')
traffic['year'] = traffic['updatetimeutc'].dt.year

# 2. Crear copias de los datasets completos
traffic_all = traffic.copy()
eco_all = eco.copy()

# 3. Mostrar los años únicos encontrados en traffic
traffic_all['year'].unique()
```

```python
# Filtrar registros del 2024
traffic_2024 = traffic[traffic['year'] == 2024].copy()
eco_2024 = eco[eco['year'] == 2024].copy()
# Vista rápida de Traffic 2024
traffic_2024.head()

# Vista rápida de Eco 2024
eco_2024.head()
```

```python
# Calcular los  promedios de trafico por ciudad, país y año
metrics = [
    'jamsdelay', 
    'trafficindexlive', 
    'jamslengthinkms', 
    'jamscount', 
    'minsdelay', 
    'traveltimeliveper10kmsmins', 
    'traveltimehistoricper10kmsmins'
]
traffic_city_year_2024 = traffic_2024.groupby(['city', 'country', 'year'])[metrics].agg('mean').reset_index()

# 3. Mostrar las primeras filas del nuevo dataset consolidado
traffic_city_year_2024.head()
```

```python
traffic_city_year_2024.sort_values(["jamsdelay"], ascending=False)
```

```python
# Seleccionar columnas clave de tráfico y economía
# 1. Seleccionar columnas clave de tráfico (usando nombres detectados)
traffic_subset = traffic_city_year_2024[['city', 'year', 'jamsdelay', 'trafficindexlive', 'minsdelay']].copy()

# 2. Seleccionar columnas clave de economía
# Asegúrate de usar los nombres que estandarizamos (gdp, unemployment, population)
eco_subset = eco_2024[['city', 'year', 'city_gdp_per_capita', 'unemployment_pct', 'population']].copy()

# 3. Unir ambos DataFrames (Inner Join) por ciudad y año
merged = pd.merge(traffic_subset, eco_subset, on=['city', 'year'], how='inner')

# 4. Mostrar el resultado final
merged.head()
```

```python
# 1. Boxplot para observar la distribución de JamsDelay (Tráfico)
plt.figure(figsize=(10, 6))
# Se agrega showmeans=True para identificar el promedio aritmético
sns.boxplot(x=merged['jamsdelay'], color='skyblue', showmeans=True,
            meanprops={"marker":"o", "markerfacecolor":"white", "markeredgecolor":"black", "markersize":"10"})
plt.title('Distribución de Retrasos por Congestión (Jams Delay) - 2024')
plt.xlabel('Minutos de Retraso Promedio')
plt.savefig('boxplot_jamsdelay.png')
plt.show()
```

```python
# 2. Histograma para observar la distribución del PIB per cápita (Economía)
plt.figure(figsize=(10, 6))
sns.histplot(merged['city_gdp_per_capita'], bins=10, kde=True, color='seagreen')
plt.title('Distribución del PIB per cápita por Ciudad - 2024')
plt.xlabel('PIB per cápita (USD)')
plt.ylabel('Frecuencia (N° de Ciudades)')
plt.savefig('histograma_gdp.png')
plt.show()
```

```python
# Ordenamos por PIB para facilitar la identificación de patrones
merged_sorted = merged.sort_values('city_gdp_per_capita', ascending=False)

fig, ax1 = plt.subplots(figsize=(14, 8))

# Barras para el PIB per cápita (Eje principal)
sns.barplot(x='city', y='city_gdp_per_capita', data=merged_sorted, ax=ax1, color='lightgray', label='PIB per cápita')
ax1.set_ylabel('PIB per cápita (USD)', color='gray', fontsize=12)
ax1.set_xlabel('Ciudad', fontsize=12)

# Crear un segundo eje para el retraso de tráfico
ax2 = ax1.twinx()
sns.barplot(x='city', y='jamsdelay', data=merged_sorted, ax=ax2, color='red', alpha=0.4, label='Retraso (Jams Delay)')
ax2.set_ylabel('Retraso Promedio (min)', color='red', fontsize=12)

plt.title('Comparativa: Nivel Económico vs. Congestión Vehicular por Ciudad (2024)', fontsize=14)
```

```python
# Rotar las etiquetas del eje X en 90 grados
plt.xticks(rotation=90) 

# Combinar las leyendas de ambos ejes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper right')

plt.tight_layout()
plt.savefig('comparativa_final.png')
plt.show()
```

```python
# Exportar el DataFrame final consolidado
merged.to_csv('ladb_mobility_economy_2024_clean.csv', index=False)

print("¡Archivo exportado con éxito!")
```
